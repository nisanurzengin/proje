## Comparison of Happiness Distributions Across Years (2007 vs 2023)

```python
country_counts = df.groupby('year')['Country name'].nunique()

count_2024 = df2024['Country name'].nunique()

country_counts.loc[2024] = count_2024
country_counts = country_counts.sort_index()

plt.figure(figsize=(8,5))
country_counts.plot(marker='o')

plt.title("Number of Countries Over Years (Including 2024)")
plt.xlabel("Year")
plt.ylabel("Number of Countries")

plt.grid()
plt.show()
print(df.groupby('year')['Country name'].nunique())
```

```python
df2007 = df[df['year'] == 2007].copy()
df2023 = df[df['year'] == 2023].copy()
common_countries = set(df2007['Country name']) & set(df2023['Country name'])

df2007_common = df2007[df2007['Country name'].isin(common_countries)].copy()
df2023_common = df2023[df2023['Country name'].isin(common_countries)].copy()
print(len(df2007_common), len(df2023_common))
```

## Relationship Between Features

```python
df_numeric = df.select_dtypes(include=['float64', 'int64'])
correlation_matrix = df_numeric.corr()
sns.heatmap(
    correlation_matrix,
    annot=True,
    fmt='.2f',
    cmap='coolwarm_r',
    linewidth=.7
)
plt.title('Relationship Between Features')
plt.show()
```

```python
df_numeric = df.select_dtypes(include=['float64', 'int64'])

correlations = df_numeric.corr()['Life Ladder'].drop('Life Ladder').sort_values(ascending=False)


cor_df = correlations.reset_index()
cor_df.columns = ['Feature', 'Correlation']

plt.figure(figsize=(10, 6))

sns.barplot(
    data=cor_df,
    x='Correlation',
    y='Feature',
    palette='RdYlGn'
)

plt.title('Correlation of Features with Happiness Score (Life Ladder)')
plt.xlabel('Correlation Coefficient')
plt.ylabel('Features')

plt.axvline(x=0, color='black', linestyle='--', linewidth=1)
plt.grid(axis='x', alpha=0.3)

plt.tight_layout()
plt.show()
```
